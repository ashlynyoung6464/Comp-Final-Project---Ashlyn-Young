```python
#Data Description

#Characterize the mean, meadian and standard deviation of all variables in your dataset relevant to your hypothesis.

#Organizing the data
import numpy as np

file = open("comp_final_project.csv", "r") #open the file
lines = file.readlines() #read the file

#list of ingredients
ingredient_list = []
for line in lines[30:40]:
    parts = line.strip().split(",")
    numbers = parts[0]
    ingredient_list.append(numbers)
print('Ingredient List', ingredient_list)

#intake list
intake = []
for line in lines[3:9]:
    parts = line.strip().split(",")
    numbers = parts[1]
    numbers = float(numbers)
    intake.append(numbers)
print('\nInput List:', intake)
mean = np.mean(intake)
median = np.median(intake)
std = np.std(intake)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)

#output list
output = []
for line in lines[3:9]:
    parts = line.strip().split(",")
    numbers = parts[1]
    numbers = float(numbers)
    output.append(numbers)
print('\nOutput List:', output)
mean = np.mean(output)
median = np.median(output)
std = np.std(output)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)

#list of test diets
testdiet_list = []
for line in lines[20:26]:
    parts = line.strip().split(",")
    numbers = parts[0]
    testdiet_list.append(numbers)
print('\nTest Diet List:', testdiet_list)

#list initial weights
initialweight_list = []
for line in lines[20:26]:
    parts = line.strip().split(",")
    numbers = parts[2]
    numbers = float(numbers)
    initialweight_list.append(numbers)
print('\nInitial Weight List:', initialweight_list)
mean = np.mean(initialweight_list)
median = np.median(initialweight_list)
std = np.std(initialweight_list)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)

#list final weights
finalweight_list = []
for line in lines[20:26]:
    parts = line.strip().split(",")
    numbers = parts[3]
    numbers = float(numbers)
    finalweight_list.append(numbers)
print('\nFinal Weight List:', finalweight_list)
mean = np.mean(finalweight_list)
median = np.median(finalweight_list)
std = np.std(finalweight_list)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)

#list of PS-MPs
PSMPs_list = []
for line in lines[20:26]:
    parts = line.strip().split(",")
    numbers = parts[1]
    numbers = float(numbers)
    PSMPs_list.append(numbers)
print('\nPSMPs:', PSMPs_list)
mean = np.mean(PSMPs_list)
median = np.median(PSMPs_list)
std = np.std(PSMPs_list)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)

#list of SGR values
SGR_list = []
for line in lines[20:26]:
    parts = line.strip().split(",")
    numbers = parts[9]
    numbers = float(numbers)
    SGR_list.append(numbers)
print('\nSGR:', SGR_list)
mean = np.mean(SGR_list)
median = np.median(SGR_list)
std = np.std(SGR_list)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)

#list of weight gain per day
wgday_list = []
for line in lines[20:26]:
    parts = line.strip().split(",")
    numbers = parts[4]
    numbers = float(numbers)
    wgday_list.append(numbers)
print('\nWeight gain per day:', wgday_list)
mean = np.mean(wgday_list)
median = np.median(wgday_list)
std = np.std(wgday_list)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)

#list of weight gain
wg_list = []
for line in lines[20:26]:
    parts = line.strip().split(",")
    numbers = parts[5]
    numbers = float(numbers)
    wg_list.append(numbers)
print('\nWeight gain:', wg_list)
mean = np.mean(wg_list)
median = np.median(wg_list)
std = np.std(wg_list)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
```

```python
#Correlation
#####Calculate Pearson and Spearman Correlation Coefficients between key variables relevant to your hypothesis.
#####Conduct Linear regression on the above variable sets.
#####Generate scatter plots of real data with linear regression lines overlaid


#linear regression
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

#Scatter plot of final weight of Labeo rohita fish vs polystyrene microplastics (PS-MPS%) ingested. 

#x = microplastics ingested
#y = final weight

x = np.array(intake)
y = np.array(finalweight_list)

plt.scatter(x, y)
plt.xlabel("PS-MPS% ingested")
plt.ylabel("Final Weight (g)")
plt.title("Scatter plot of final weight (g) vs polystyrene microplastics (PS-MPS%) ingested by Labeo rohita fish")

#Linear regression
LR = stats.linregress(x, y)

#Adding slope / linear regression line to scatter plot
slope_line = LR.slope * x + LR.intercept
plt.plot(x, slope_line)
plt.show()

#Pearson correlation coefficients
pearson_list = []
res = stats.pearsonr(x, y)
pearson_coef = res.statistic
pearson_list.append(pearson_coef)

#Spearman correlation coefficients
spearman_list = []
res = stats.spearmanr(x, y)
spearman_coef = res.statistic
spearman_list.append(spearman_coef)

print("R value:", LR.rvalue)
print('Pearson correlation coefficient:', pearson_list)
print('Spearman correlation coefficient:', spearman_list)
```

```python
#Dimensionality reduction

#Conduct Principle Components Analysis (CPA) using all quantitative variables in your data set for across all samples.

import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

varnames = ["PS-MPs", "Initial W", "Final W", "WG/Day", "WG", "SGR"]

#Making a matrix
x = np.column_stack([
    PSMPs_list,
    initialweight_list,
    finalweight_list,
    wgday_list,
    wg_list,
    SGR_list
])

#
#Making a matrix
x = np.column_stack([
    PSMPs_list,
    initialweight_list,
    finalweight_list,
    wgday_list,
    wg_list,
    SGR_list])

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

#Doing PCA calcs
pca = PCA(n_components=2)
scores = pca.fit_transform(x_scaled)


#Extract principle component weights and plot using a bar chart.

#Bar Chart for PCA Component 1
plt.subplot(1, 2, 1)
plt.bar(varnames, pca.components_[0])
plt.xlabel("Variables")
plt.ylabel("PCA Weight")
plt.title("PCA 1")
plt.xticks(rotation=45)
plt.tight_layout()

#Bar Chart for PCA Component 2
plt.subplot(1, 2, 2)
plt.bar(varnames, pca.components_[1])
plt.xlabel("Variables")
plt.ylabel("PCA Weight")
plt.title("PCA 2")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


#Visualize your data on PC1 and PC2. Color code data points by relevant categories.
plt.figure(figsize=(6,5))

plt.scatter(scores[0,0], scores[0,1], color="red", label="I")
plt.text(scores[0,0] + 0.05, scores[0,1] + 0.05, "I")

plt.scatter(scores[1,0], scores[1,1], color="orange", label="II")
plt.text(scores[1,0] + 0.05, scores[1,1] + 0.05, "II")

plt.scatter(scores[2,0], scores[2,1], color="yellow", label="III")
plt.text(scores[2,0] + 0.05, scores[2,1] + 0.05, "III")

plt.scatter(scores[3,0], scores[3,1], color="green", label="IV")
plt.text(scores[3,0] + 0.05, scores[3,1] + 0.05, "IV")

plt.scatter(scores[4,0], scores[4,1], color="blue", label="V")
plt.text(scores[4,0] + 0.05, scores[4,1] + 0.05, "V")

plt.scatter(scores[5,0], scores[5,1], color="purple", label="VI")
plt.text(scores[5,0] + 0.05, scores[5,1] + 0.05, "VI")

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Plot of PC2 vs PC1")
plt.grid(True)
plt.legend(title="Test Diets")
plt.tight_layout()
plt.show()

#Printing PCA results
print("PCA Components:",pca.components_)
print("Explained Variance Ratio (PC1, PC2) :", pca.explained_variance_ratio_)
```
